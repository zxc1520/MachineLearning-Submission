## Pengerjaan

```python
df2 = pd.read_csv("data/insurance.csv")
df2.head()
```

```python
new_labels = {
     'male': 1,
     'female': 0
}

df2['sex'] = df2['sex'].map(new_labels)
df2.head()
```

```python
new_labels_region = {
     'southwest': 0,
     'southeast': 1,
     'northwest' : 2,
     'northeast' : 3
}

df2['region'] = df2['region'].map(new_labels_region)
df2.head()
```

```python
new_labels_smoker = {
     'yes': 1,
     'no': 0
}

df2['smoker'] = df2['smoker'].map(new_labels_smoker)
df2.head()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor


X = df2[['age', 'sex', 'bmi', 'children', 'smoker', 'region']]

vif_data = pd.DataFrame()
vif_data["feature"] = X.columns

vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]

print(vif_data)
```

```python
df_copy = df2.copy()
df_copy = df_copy.drop(df_copy.columns[0], axis=1)
df_copy = df_copy.drop(df_copy.columns[2], axis=1)

df_copy.head()
```

```python
df3 = df_copy.copy()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

X = df3.iloc[:, :-1]
y = df3.iloc[:, 4]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.2, random_state=50)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)
```

```python
con = np.concatenate((y_test, y_pred))
con
```

```python
from sklearn.metrics import r2_score

r_square = r2_score(y_test, y_pred)
r_square
```
